### Diagramas de Caja y Datos Anómalos

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Creación de Diagramas de Caja (BoxPlots)

* Si nosotros sacaramos un promedio de estos datos realmente no tendría mucho sentido, ya que las edades son muy disparejas, lo que puede afectar la media de la edad.
* Boxplot trabaja con los datos ordenados, para este caso ya lo tenemos.
* Boxplot nos muestra el minimo y el máximo con las 2 lineas.
* La linea naranja representa la mediana de nuestros datos.
* Nos muestra los cuartiles, es decir parte nuestros datos de 4 en 4.
* Podemos ver que cuando agregamos eso 4 cuatros, el primer cuartil se hizo muy pequeño.

```python
# [4,4,4,4, 17, 29, 33, 50, 51, 61, 71, 85, 92, 99
edades = np.array(
    [4, 17, 29, 33, 50, 51, 61, 71, 85, 92, 99]
)

plt.boxplot(edades)
plt.grid(True)
plt.yticks(np.arange(0,110,5))
plt.show()
```

### Cuantiles

* Así se construyó nuestro boxplot.
* En caso se calculó lo siguiente:
  * Sacamos el valor minimo.
  * Calculamos el primer,segundo y tercel cuartil.
  * El ultimo cuartil por default sería del tercel cuartil al valor máximo.

```python
(
    np.min(edades),
    np.quantile(edades,0.25),
    np.median(edades),
    np.quantile(edades, 0.75),
    np.max(edades)
)
```

### Datos Anormales / Atípicos / Outliers

### Prueba de Tukey

* Podemos ver que con el mismo código, pero con diferentes datos.
* Vemos que los bigotes del boxplot ahora no tocan los ultimos ni primeros valores.
* Los datos anómalos son los que marca con el circulito.
* Python por default hace esta prueba, asumiendo que los datos siguen una prueba normal.

```python
edades = np.array(
    [6,41,51,52,54,57,60,61,65,95,100]
)

plt.grid(True)
plt.yticks(np.arange(0,110,5))
plt.boxplot(edades)
plt.show()
```
