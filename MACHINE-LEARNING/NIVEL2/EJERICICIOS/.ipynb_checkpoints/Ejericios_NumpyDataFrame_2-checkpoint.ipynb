### Ejericio 1

Analiza los gastos personales de una persona durante un mes.

* Calcular el gasto total.
* Calcular el gasto por categoría.
* Encontrar el día con mayor gasto.
* Visualizar los gastos por categoría con plot(kind="bar").

```python
datos = {
    "fecha": pd.date_range("2024-04-01", periods=10),
    "categoría": ["Comida", "Transporte", "Alquiler", "Comida", "Ocio", "Comida", "Transporte", "Salud", "Ocio", "Comida"],
    "monto": [12.5, 3.2, 500.0, 15.0, 25.0, 10.0, 3.5, 40.0, 20.0, 13.0]
}
gastos = pd.DataFrame(datos)

#Cantidad total de gastos
monto = gastos["monto"].sum()
print("\nGasto total:", monto)

# Gasto total por categoría, ponemos el set para que no salga el tipo de dato
monto_cat = gastos.groupby("categoría")["monto"].sum().reset_index()
print("\nMonto por Categoría:", monto_cat)

# Encontramos el día con mayor gasto, tener cuidado ya que es una fecha.
max_dia = gastos.groupby("fecha")["monto"].sum().idxmax()
print("\nDía máximo:", max_dia)
```

```python
# Graficar
monto_cat.plot(kind="bar", color="skyblue", figsize=(8, 5))

# Establecemos las etiquetas
plt.title("Gastos totales por categoría")
plt.xlabel("Categoría")
plt.ylabel("Monto gastado")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

* El parámetro `as_index=False` se utiliza en `groupby()` de pandas para indicar que la columna por la que agrupas no se convierta en el índice del resultado.
* Cuando haces un `.groupby()` sin especificar `as_index=False`, pandas usa la(s) columna(s) de agrupación como índice del DataFrame resultante.
*  ¿Cuándo usarlo?
    * Cuando quieras gráficos más sencillos (plot(), bar()).
    * Cuando planeas guardar a CSV sin complicar el índice.
    * Cuando vas a fusionar (merge) con otro DataFrame.

```python
# Agrupar y convertir a DataFrame para control total
montos = gastos.groupby("categoría", as_index=False)["monto"].sum()

# Graficar con nombres en eje x
plt.figure(figsize=(8, 5))
plt.bar(montos["categoría"], montos["monto"], color="mediumseagreen")
plt.title("Gastos totales por categoría")
plt.xlabel("Categoría")
plt.ylabel("Monto gastado")
plt.xticks(rotation=45)  # Rota las etiquetas si son muchas o largas
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

### Ejercicio 2

Comparar el rendimiento promedio entre diferentes cursos.

1. Calcular el promedio de nota por curso.
2. Ver qué curso tuvo el mejor rendimiento.
3. Mostrar cuántos estudiantes hay por curso.
4. Graficar la media de cada curso.

***NOTAS***

```python
data = {
    "nombre": ["Ana", "Luis", "María", "Carlos", "Elena", "Pablo", "Lucía", "Andrés"],
    "curso": ["A", "A", "B", "B", "C", "C", "A", "B"],
    "nota_final": [8.5, 6.0, 9.0, 7.0, 5.5, 6.5, 8.0, 4.5]
}
notas = pd.DataFrame(data)

# Promedios por grupo
max_means = notas.groupby("curso")["nota_final"].mean().reset_index()
print("\nPromedios por Materia:\n", max_means)

# Ver el curso con mejor rendimiento
max_curso = max_means.loc[max_means["nota_final"].idxmax()]
print("\nCurso:\n", max_curso)

# Cuantos estudiantes hay por grupo
conteo = notas["curso"].value_counts().reset_index()
print("\nCantidad de estudiantes por curso:")
print(conteo)

# Graficar los promedios de curso
plt.figure(figsize=(8, 5))
plt.bar(max_means["curso"], max_means["nota_final"], color="mediumseagreen")
plt.title("Promedios")
plt.xlabel("Materia")
plt.ylabel("Promedio")
plt.xticks(rotation=45)  # Rota las etiquetas si son muchas o largas
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

### Ejercicio 3

Trabajar con series temporales de precios de acciones.

* Calcular el cambio diario (`diff()`).
* Calcular la media móvil de 3 días (`rolling().mean()`).
* Detectar si hubo algún día con caída mayor a 2 unidades.
* Graficar la serie original y la media móvil.

***NOTAS***
* `diff` se utiliza para calcular la diferencia entre elementos consecutivos de una matriz o DataFrame.
* En NumPy, `numpy.diff()` calcula la diferencia de primer orden.
* En Pandas, `DataFrame.diff()` calcula la diferencia entre cada fila consecutiva. Podemos expecificar la cantidad de filas.
* `df_diff = acciones.diff()`: es este caso funciono, porque nuestro dataframe es de solo 2 columnas, pero tenemos que especificar.
* Una ***media móvil de 3 días (o 3-day moving average)*** es un tipo de promedio que suaviza una serie de datos calculando la media de cada punto con los dos anteriores.
* Tomas cada grupo de 3 días consecutivos y calculas el promedio de sus valores. Así, ves la tendencia general sin que los altibajos diarios afecten tanto.
* La media móvil de 3 días se calcula así:
    * Día 3: (10 + 12 + 14) / 3 = 12.0
    * Día 4: (12 + 14 + 13) / 3 = 13.0
    * Día 5: (14 + 13 + 15) / 3 = 14.0
    * No hay valores en días 1 y 2 porque necesitas al menos 3 días para calcular la primera media.
* ¿Para qué se usa?
    * Eliminar ruido (variaciones diarias pequeñas)
    * Detectar tendencias (¿sube? ¿baja?)
    * Muy común en series de tiempo: precios, clima, ventas, etc.
* `rolling().mean()` es una función de Pandas en Python que calcula la media móvil de una serie de datos.
* Calcula la media de los valores dentro de una ***"ventana"*** móvil que se desplaza a través de la serie. Esta función es útil para suavizar datos, identificar tendencias y realizar análisis de series temporales.
* `rolling()`:
    * Este método crea un objeto "rodante" que define la ventana móvil y cómo se desplaza a través de la serie. Se le puede pasar un parámetro window
      que especifica el tamaño de la ventana (número de valores que se incluyen en el cálculo de la media). 

```python
# Simular precios diarios de una acción durante 10 días
acciones = pd.DataFrame({
    "fecha": pd.date_range("2024-01-01", periods=10),
    "precio": [100 + np.random.randn() * 2 for x in range(10)]
})
# Mostramos las acciones
print("\nAcciones:\n", acciones)

# Calcula la diferencia de primer orden (filas), en este caso de 1 en 1.
acciones["cambio_diario"] = acciones["precio"].diff()
print("\nDiferencia entre Precios:\n", acciones)

# Media movil.
acciones["media_movil_3d"] = acciones["precio"].rolling(window=3).mean()
print("\n Media Móvil de 3 días:\n", acciones)

# Días con caida de2 unidades
caidas = acciones[acciones["cambio_diario"] < -2]
print("Días con caída mayor a 2 unidades:")
print(caidas[["fecha", "precio", "cambio_diario"]])

# Gráfica
plt.figure(figsize=(8, 5))
# Establecemos los 2 grafos.
plt.plot(acciones["fecha"], acciones["precio"], label="Precio original", marker="o")
plt.plot(acciones["fecha"], acciones["media_movil_3d"], label="Media móvil 3 días", linestyle="--", color="orange")

plt.title("Precio diario vs. media móvil")
plt.xlabel("Fecha")
plt.ylabel("Precio")
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Ejercicio 4

Usar NumPy para análisis básico de datos numéricos.
* Calcular el promedio de cada sensor.
* Calcular la temperatura máxima por día.
* Estándarizar los datos (restar media y dividir por desviación).
* Detectar valores mayores a 22°C.

```python
# Simular 5 sensores que miden temperatura durante 6 días
data = np.array([
    [20.1, 21.0, 20.8, 21.5, 22.0, 20.5],
    [19.8, 20.3, 20.5, 21.0, 21.2, 20.7],
    [21.0, 21.5, 21.3, 21.8, 22.1, 21.0],
    [20.0, 20.5, 20.2, 20.9, 21.3, 20.8],
    [19.5, 20.0, 19.8, 20.3, 20.9, 20.2]
])

# Calcular el promedio por filas (axis=1), espeficamos que sea por fila
promedio_por_filas = np.mean(data, axis=1)
print("\nPromedio por cada sensor:",promedio_por_filas)

# Calcular la máxima temperatura por día
maxima_temp = np.max(data, axis=0)
print("\nMáxima temp por cada día:",maxima_temp)

# Estandarización de datos (usemos calculos vectoriales)
std = (data - np.mean(data ))/ np.std(data )
print("\nEstandarización de Datos:\n", std)

# Valores mayor a 22.
print(data[np.where(data > 22)])
print(data[data > 22])
valores_altos = data > 22.0
print("¿Valores > 22°C?:\n", valores_altos)
print("Valores detectados:", data[valores_altos])
```

### Ejercicio 5

Combinar estructuras de control de Python, arrays de NumPy y DataFrames para analizar temperaturas.

1. Agrega una columna con el promedio diario.
2. Encuentra qué región tuvo la mayor temperatura el miércoles.
3. Crea una nueva columna que indique si el promedio del día fue > 25°C.
4. Convierte las temperaturas de °C a °F usando NumPy (temp * 9/5 + 32).

***NOTAS***
* `np.random.seed(0)` : Fija la semilla aleatoria para que los números generados sean siempre los mismos.
* `np.random.randint(15, 35, size=(5, 3))`: Genera una matriz de 5 filas y 3 columnas con enteros aleatorios entre 15 y 34 (el 35 no está incluido).


```python
regiones = ["Norte", "Centro", "Sur"]
días = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes"]

# Crear una matriz de temperaturas aleatorias por región y día
np.random.seed(0)
temp = np.random.randint(15, 35, size=(5, 3))

df = pd.DataFrame(temp, index=días, columns=regiones)

df["Promedio"] = df[["Norte", "Centro","Sur"]].mean(axis=1)

print("\nTemperatura máxima de cada día:\n",df.max(axis=1).reset_index())
print("\nTemperatura máxima del Miercoles: ",df.loc["Miércoles"].max())

#Columna indique si el promedio fue mayor a 25.
df["More 25?"] = df["Promedio"] > 25
print("\nTemperaturas mayores a 25:\n", df)

# df.loc[:, ["Norte", "Centro","Sur",]] = df[["Norte", "Centro","Sur"]] * 9/5 + 32

# print("\nTemperaturas en Farenheit:\n", df)
```
